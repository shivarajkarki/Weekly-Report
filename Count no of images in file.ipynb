```python
import os

classes = ['Dry Edge', 'Edge Tear', 'Floating Lens', 'Folded Lens',
           'Gapped Edge', 'In Lens Tear',
           'Multiple Lens', 'No Lens', 'No Shell', 'Particles']
```

```python
count = [0]*len(classes)
```

```python
for r, d, f in os.walk(os.getcwd()):
    if classes[0] in r:
        print(r)
        for file in f:
            if '.bmp' in file:
                print(file)
                count[0] += 1

    if classes[1] in r:
        print(r)
        for file in f:
            if '.bmp' in file:
                print(file)
                count[1] += 1
        
    
    if classes[2] in r:
        print(r)
        for file in f:
            if '.bmp' in file:
                print(file)
                count[2] += 1
        
    
    if classes[3] in r:
        print(r)
        for file in f:
            if '.bmp' in file:
                print(file)
                count[3] += 1
                
    if classes[4] in r:
        print(r)
        for file in f:
            if '.bmp' in file:
                print(file)
                count[4] += 1
    
    if classes[5] in r:
        print(r)
        for file in f:
            if '.bmp' in file:
                print(file)
                count[5] += 1
    
    if classes[6] in r:
        print(r)
        for file in f:
            if '.bmp' in file:
                print(file)
                count[6] += 1
                
    if classes[7] in r:
        print(r)
        for file in f:
            if '.bmp' in file:
                print(file)
                count[7] += 1
                
    if classes[8] in r:
        print(r)
        for file in f:
            if '.bmp' in file:
                print(file)
                count[8] += 1
                
    if classes[9] in r:
        print(r)
        for file in f:
            if '.bmp' in file:
                print(file)
                count[9] += 1
        
print(count)
```

```python
for i,num in enumerate(count):
    print('We have {} number of images of class {}'.format(num, classes[i]))
```

```python
import matplotlib.pyplot as plt
```

```python
fig1, ax1 = plt.subplots(figsize=(8,8))
ax1.pie(count, labels=classes,
        autopct=lambda p: '{:.2f}%({:.0f})'.format(p,(p/100)*sum(count)),
        shadow=True, startangle=90, radius = 5)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.legend(loc= 'upper right')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
