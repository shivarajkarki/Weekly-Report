```python
import numpy as np
#import cv2
import matplotlib.pyplot as plt
import glob
import cv2 as cv
import imutils
import os
import time
```

```python
"""
This function Flips, Translates and rotates each image,
Done using list comprehension.
"""

def Aug_list_comprehension(input_list):
    
    cv_img = input_list
    
    """ 0, for flipping the image around the x-axis (horizontal flipping);
        > 0, for flipping around the y-axis (verticle flipping);"""
    
    flip = [0, 1] 
    
    """Linear translation """
    
    pixels_lapse = 10

    M0 = np.float32([[1,0,0],[0,1,0]]) #linear translation... lower right corner
    M1 = np.float32([[1,0,pixels_lapse],[0,1,pixels_lapse]]) #linear translation... lower right corner
    M2 = np.float32([[1,0,-pixels_lapse],[0,1,pixels_lapse]]) #linear translation... lower left corner
    M3 = np.float32([[1,0,pixels_lapse],[0,1,-pixels_lapse]]) #linear translation... upper right corner
    M4 = np.float32([[1,0,-pixels_lapse],[0,1,-pixels_lapse]]) #linear translation... upper left corner

    M = [M0, M1, M2, M3, M4]
    
    #read dimension of image
    img_height, img_width = cv_img[0].shape

    
    #flip the image
    cv_img_flip = [cv.flip(item,f) for item in cv_img for f in flip]
    
    #translate the flipped image
    cv_img_flip_translate = [cv.warpAffine(item, m, (img_width, img_height)) for item in cv_img_flip for m in M]
    
    #rotate the flipped translated image
    cv_img_flip_translate_rotate = [imutils.rotate(item, angle) for item in cv_img_flip_translate 
                              for angle in np.arange(0,360, 15)]
    
    return cv_img_flip_translate_rotate
```

```python
directories=[d for d in os.listdir(os.getcwd()) if os.path.isdir(d)]

path = [item for item in directories[1:]]
```

```python
"""
Reads the image from folder and calls function to augment each image, and write back to same folder
print statements are used to debug during run time
"""

def Aug_images(each_path):
    
    
    print('Inside: ', each_path, '...')
    
    cv_image = []
    
    #read images from given folder

    for img in glob.glob(each_path+'/*.bmp'):
        n = cv.imread(img,0) #read gray scale image
        cv_image.append(n)
    
    #.....
    print('each Image type:',type(cv_image[0]))
    print('Total no of images: {}'.format(len(cv_image)))
    print('....Images are taken from ', each_path, 'and read')
    print('.\n..\n...Augmenting Images')
    #.....
    
    #to find out time taken to augment images
    start = time.time()
    #....
    
    cv_aug_list_compre = Aug_list_comprehension(cv_image)
    
    #....
    print('...\tAugmentation Done in time: ',time.time()-start,' sec...!!!')
    print('Writing',len(cv_aug_list_compre),' no of images to file...')
    print('....\n......\n')
    #....
    
    #writing image back to folder
    for i, image in enumerate(cv_aug_list_compre):
        result = cv.imwrite(each_path+"/img"+str(i)+".bmp", image)
        
    print("...Finished Writing Images....\n\n")
        
```
